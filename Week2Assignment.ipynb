| customer | chocolate | sprinkles | filled | glazed |
|----------|-----------|-----------|--------|--------|
| A        | ?         | ?         | ?      | ?      |
| B        | ?         | ?         | ?      | ?      |
| C        | ?         | ?         | ?      | ?      |


After you have completed your code, evaluate it for performance bottlenecks and describe how you would improve the code.  

*Bonus Question:* If the two starting tables were in a relational database or Hadoop cluster and each had a 100 million rows, how might your approach change?

````python
import pandas as pd
import numpy as np

tableA = [(100, 'chocolate, sprinkles'),
         (101, 'chocolate, filled'),
         (102, 'glazed')]
labels = ['product', 'tags']
dfA = pd.DataFrame.from_records(tableA, columns=labels)

tableB = [('A', 100),
           ('A', 101),
           ('B', 101),
           ('C', 100),
           ('C', 102),
           ('B', 101),
           ('A', 100),
           ('C', 102)]
labels = ['customer', 'product']
dfB = pd.DataFrame.from_records(tableB, columns=labels)
````

````python

dfA
````

````python
#Answer to question two goes here

# dummy variables. a, b, c -> letters,
# letters: [a, b, c]
# letters_a :[1, 0, 0]
# letters_b: [0, 1, 0]
# letters_c: [0,0,1]


data= pd.merge(dfA, dfB, on='product', how='outer', suffixes=('_x', '_y'))
````

````python
data
````

````python
new = data["tags"].str.split(",", n = 1, expand = True) 
````

````python
data['x']=new[0]
data['y']=new[1]
data=data.drop('tags',1)
````

````python

data1 = data.filter(['customer','y','product'], axis=1)
````

````python
data
````

````python
data1.columns=['customer','x','product']
````

````python
data2=data.append(data1,sort=True)
````

````python
data3=data2.drop('y',1)
````

````python
dummies = pd.get_dummies(data3['x'])
````

````python
dummies['customer']=data3['customer']
````

````python
dummies.columns
````

````python
final_result=dummies.groupby('customer')[['chocolate',' sprinkles',' filled','glazed']].sum().reset_index()
````
