```python
import re
def clean_text(text):
    #allow only alphabetes
    text=re.sub(r'[^a-zA-Z\']',' ',text)
    #remove unicode characters
    text=re.sub(r'[^\x00-\x7F]+','',text)
    #convert to lower case
    text.lower()
    return text
```

```python
df['clean_text']=df.tweet.apply(lambda x:clean_text(x))
```

```python
from wordcloud import STOPWORDS
```

```python
print(STOPWORDS)
```

```python
def gen_freq(text):
    word_list=[]
    for tw_word in text.split():
        word_list.extend(tw_word)
    word_freq=pd.Series(word_list).value_counts()
    word_freq=word_freq.drop(STOPWORDS, errors='ignore')
    return word_freq
    
```

```python
def gen_freq1(text):
    #Will store the list of words
    word_list = []

    #Loop over all the tweets and extract words into word_list
    for tw_words in text.split():
        word_list.extend(tw_words)

    #Create word frequencies using word_list
    word_freq = pd.Series(word_list).value_counts()
    
    #Drop the stopwords during the frequency calculation
    word_freq = word_freq.drop(STOPWORDS, errors='ignore')
    
    return word_freq
```

```python
#check whether a negation word present in text
def any_neg(words):
    for word in words:
        if word in ['n','no','not','non'] or re.search(r"\wn't",word):
            return 1
        else:
            return 0
        
```

```python
#check one of the 100 rare words present in text
def any_rare(words,rare_100):
    for word in words:
        if word in rare_100:
            return 1
        else:
            return 0
    
    
```

```python
#check whether prompt wordes are present
def is_question(words):
    for word in words:
        if word in ['what','why','how','when','who']:
            return 1
        else:
            return 0
```

```python
x=df.clean_text[-100:]
```

```python
word_freq=gen_freq(df.clean_text.str)
rare_100=word_freq[-100:]
df['word_count']=df.clean_text.str.split().apply(lambda x: len(x))
df['any_neg']=df.clean_text.str.split().apply(lambda x:any_neg(x))
df['is_question']=df.clean_text.str.split().apply(lambda x:is_question(x))
df['any_rare']=df.clean_text.str.split().apply(lambda x:any_rare(x,rare_100))
df['char_count']=df.clean_text.apply(lambda x:len(x))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=123)
```
