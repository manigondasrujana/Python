```python
import pandas as pd

tableA = [(100, 'chocolate, sprinkles'),
         (101, 'chocolate, filled'),
         (102, 'glazed')]
labels = ['product', 'tags']
dfA = pd.DataFrame.from_records(tableA, columns=labels)

tableB = [('A', 100),
           ('A', 101),
           ('B', 101),
           ('C', 100),
           ('C', 102),
           ('B', 101),
           ('A', 100),
           ('C', 102)]
labels = ['customer', 'product']
dfB = pd.DataFrame.from_records(tableB, columns=labels)
```

```python
dfA
```

```python
dfB
```

```python
data= pd.merge(dfA, dfB, on='product', how='outer', suffixes=('_x', '_y'))
```

```python
data.head()
```

```python
new = data["tags"].str.split(",", n = 1, expand = True) 
```

```python
data['x']=new[0]
```

```python
data['y']=new[1]
```

```python
data
```

```python
data1=data
```

```python
data2=data1.drop('tags',1)
```

```python
data2
```

```python
data3 = data2.filter(['customer','y','product'], axis=1)
```

```python
data3
```

```python
data3.columns=['customer','x','product']
```

```python
data4=data2.append(data3,sort=True)
```

```python
data4
```

```python
data4.drop('y',1)
```

```python
dummies = pd.get_dummies(data4['x'])
```

```python
dummies.head()
```

```python
dummies['customer']=data4['customer']
```
